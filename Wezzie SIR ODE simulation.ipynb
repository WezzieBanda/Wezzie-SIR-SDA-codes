```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt #import libraries
from matplotlib import rc
```

```python
import matplotlib.pyplot as plt
import numpy as np

N=60041996 # population size
def dU_dt(U,t): #define function 
    return [-(beta*U[0]*U[1])/N, (beta*U[0]*U[1])/N - gamma*U[1], gamma*U[1]] #SIR ODEs

U0=[60041996, 1, 0]  #initial condtions showing N,I,R

#beta, gamma= 1.66*10**(-7) ,0.11

beta, gamma= 0.132, 0.068 #initialize beta and gamma
ts=np.arange(0, 450, 0.001) #time interval

Us=odeint(dU_dt, U0, ts) #solve te ODE

S, I, R=Us[:,0], Us[:,1], Us[:, 2]


fig=plt.subplots(figsize=(8,4))
plt.plot(ts, S, linewidth=2.5, label="Suspectibles") #Plot Susceptible
plt.plot(ts, I, linewidth=2.5, label="Infectious Population") #plot infectious
#plt.plot(Daily_100, linewidth=2.5, label="cases")
plt.plot(ts, R, linewidth=2.5, label="Recovered") #plot recovered 
#plt.title("Susceptible,Infectious and Recovered Populations. Beta=1.66e-07, Gamma=0.11")
plt.xlabel("Time in days") #plot x axis
plt.ylabel("Infectious Population") #y axis
txt=",beta=1.66^10^(-7), gamma=0.11"
plt.legend(fontsize=15)
plt.grid() #show grid
plt.show() #sho wgraph
```
